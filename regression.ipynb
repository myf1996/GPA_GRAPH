```python
import pandas as pd
import numpy as np
from sklearn import metrics
from sklearn.neighbors import KNeighborsRegressor
from sklearn.ensemble import AdaBoostRegressor
from sklearn.linear_model import LinearRegression
from sklearn.preprocessing import StandardScaler
from sklearn.preprocessing import OneHotEncoder
from sklearn.model_selection import train_test_split
```

```python
df=pd.read_csv("/content/sample_data/datasets.csv")
```

```python
df=df.fillna(0)
df
```

```python
x=df.drop(['AVG GPA'], axis = 1)
y=df['AVG GPA']
```

```python
#Recursive Feature Elimination
x=df[['SEMESTER','AGE','ADDRESS','FAMILY SIZE','HEALTH','FAMILY SUPPORT','STUDY TIME','NO OF WITHDRAWL']]
```

```python
#Chi-Squared
x=df[['SEMESTER','ADDRESS','FAMILY SIZE','HEALTH','FAMILY SUPPORT','STUDY TIME','NO OF WITHDRAWL']]
```

```python
#Pearson Correlation
x=df[['SEMESTER','AGE','ADDRESS','FAMILY SIZE','FAMILY SUPPORT','STUDY TIME','NO OF WITHDRAWL']]
```

```python
#lasso: SelectFromModel l2
x=df[['AGE','FAMILY SIZE','FAMILY SUPPORT','STUDY TIME','NO OF WITHDRAWL']]
```

```python
#tree-based: SelectFromModel
x=df[['SEMESTER','AGE','INTERNET','FREETIME','HEALTH','NO OF ABSENCE','NO OF WITHDRAWL','CURRENT CGPA']]
```

```python
#lightGBM
x=df[['SEMESTER','AGE','TRAVEL TIME','INTERNET','HEALTH','NO OF ABSENCE','NO OF WITHDRAWL','CURRENT CGPA']]
```

```python
myData=[[8.0,22.0,2.0,3.0,2.0,3.0,0.0,3.2]]
```

# Linear Regression

```python
x_train, x_test, y_train, y_test = train_test_split(x, y, test_size=0.2,random_state=30)
linear_model = LinearRegression()
model = linear_model.fit(x_train, y_train)
predictions = linear_model.predict(x_test)
accuracy_scale = metrics.r2_score(y_test, predictions)
print('Accuracy Scale (0 - 4): {}'.format(accuracy_scale))
```

```python
predictions = linear_model.predict(myData)
predictions
```

# K Nearest Neighbors

```python
knn = KNeighborsRegressor(n_neighbors=16)
x_train, x_test, y_train, y_test = train_test_split(x, y, test_size=0.2, random_state=30)
knn.fit(x_train, y_train)
pred = knn.predict(x_test)
acc = metrics.r2_score(y_test, pred)
print("Accuracy: {}".format(acc))
```

```python
predictions = knn.predict(myData)
predictions
```

# ADA Boost

```python
adaBoost = AdaBoostRegressor(n_estimators=50, learning_rate=1)
x_train, x_test, y_train, y_test = train_test_split(x, y, test_size=0.2, random_state=30)
adaBoost.fit(x_train, y_train)
pred=adaBoost.predict(x_test)
accuracy = metrics.r2_score(y_test, pred)
print('Accuracy: {}'.format(accuracy))
```

```python
predictions = adaBoost.predict(myData)
predictions
```

# Neural Networks

```python
X = df.iloc[:,:18].values
y = df.iloc[:,18:19].values
```

```python
#Normalizing the data
sc = StandardScaler()
X = sc.fit_transform(X)
```

```python
ohe = OneHotEncoder()
y = ohe.fit_transform(y).toarray()
```

```python
X_train,X_test,y_train,y_test = train_test_split(X,y,test_size = 0.3)
```

```python
from keras.models import Sequential
from keras.layers import Dense

model = Sequential()
model.add(Dense(32, input_dim=18, activation='relu'))
model.add(Dense(28, activation='relu'))
model.add(Dense(23, activation='relu'))
model.add(Dense(19, activation='relu'))
model.add(Dense(15, activation='relu'))
model.add(Dense(11, activation='softmax'))
```

```python
model.compile(loss='categorical_crossentropy', optimizer='adam', metrics=['accuracy'])
```

```python
history = model.fit(X_train, y_train, epochs=100, batch_size=128)
```
